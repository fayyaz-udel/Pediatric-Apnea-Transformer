```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 14})
fig = plt.figure()

####### Noisy Modality ########
x = np.array(['10', "20", "30", "40", "50"])[::-1]

y_nch = np.array([79.0, 86.4, 89.0, 90.8, 90.9])[::-1]
e_nch = np.array([2.8, 1.2, 0.8, 0.9, 0.9])[::-1]
plt.errorbar(x, y_nch, e_nch, label='CNN', marker = 'o', alpha=0.75)

y_nch = np.array([90.1, 91.0, 91.6, 92.3, 93.2])[::-1]
e_nch = np.array([1.1, 1.0, 0.9, 1.0, 1.0])[::-1]
plt.errorbar(x, y_nch, e_nch, label='CNN+LSTM', marker = 'o', alpha=0.75)


y_nch = np.array([84.9, 85.6, 87.5, 89.4, 89.6])[::-1]
e_nch = np.array([0.3, 0.3, 0.5, 0.5, 0.4])[::-1]
plt.errorbar(x, y_nch, e_nch, label='Transformer', marker = 'o', alpha=0.75)


y_nch = np.array([86.9, 91.6, 93.2, 93.4, 93.4])[::-1]
e_nch = np.array([2.5, 0.7, 0.3, 0.3, 0.3])[::-1]
plt.errorbar(x, y_nch, e_nch, label='Ours', marker = 'o', alpha=0.75)


plt.ylim([50,100])
plt.xlabel('Signal to Noise Ratio (dB)')
plt.ylabel('AUROC')
plt.legend(loc='lower right')
fig.set_size_inches(6, 4)
plt.savefig('.\\img\\noisy.svg', format="svg")
```
