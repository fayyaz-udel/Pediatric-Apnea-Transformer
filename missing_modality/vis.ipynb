```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 13})
fig = plt.figure()

####### Noisy Modality ########
x = np.array(['10', "20", "30", "40", "50"])[::-1]

y_nch = np.array([58.4, 79.0, 88.3, 88.8, 88.8])[::-1]
e_nch = np.array([2.6, 1.4, 1.2, 1.2, 1.3])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='CNN', marker = '', alpha=0.75)

y_nch = np.array([56.7, 69.4, 86.6, 88.5, 88.5])[::-1]
e_nch = np.array([4.6, 8.0, 1.5, 1.1, 1.4])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='CNN+LSTM', marker = '', alpha=0.75)


y_nch = np.array([84.8, 85.6, 87.4, 89.3, 89.6])[::-1]
e_nch = np.array([0.2, 0.2, 0.4, 0.5, 0.4])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='Transformer', marker = '', alpha=0.75)


y_nch = np.array([87.4, 91.8, 93.3, 93.4, 93.4])[::-1]
e_nch = np.array([1.6, 0.7, 0.5, 0.5, 0.5])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='Ours', marker = '', alpha=0.75)


plt.ylim([50,100])
plt.xlabel('Signal to Noise Ratio (dB)')
plt.ylabel('AUROC')
plt.legend(loc='lower left')
fig.set_size_inches(6, 4)
plt.savefig('.\\img\\noisy.svg', format="svg")
```
