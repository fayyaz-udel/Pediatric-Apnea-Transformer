```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()

####### Missing Modality ########
x = np.array(['0%', "10%", "20%", "30%", "40%", "50%"])

# y_chat = np.array([78.42, 78.18, 79.31, 82.56, 77.19, 78.67])
# plt.plot(x, y_chat, label='SE-MSCNN', marker = 'o', alpha=0.75)

y_nch = np.array([84.62, 86.54, 88.48, 89.92, 87.31, 90.08])
plt.plot(x, y_nch, label='CNN+LSTM', marker = 'o', alpha=0.75)

y_nch = np.array([81.71, 81.85, 82.54, 85.09, 81.99, 84.66])
plt.plot(x, y_nch, label='CNN', marker = 'o', alpha=0.75)

y_nch = np.array([84.77, 86.45, 87.08, 89.50, 86.80, 89.63])
plt.plot(x, y_nch, label='Transformer', marker = 'o', alpha=0.75)

y_nch = np.array([86.44, 88.23, 89.28, 90.66, 88.14, 90.57])
plt.plot(x, y_nch, label='Ours', marker = 'o', alpha=0.75)

plt.ylim([50,100])
plt.xlabel('Missingness Ratio')
plt.ylabel('AUROC')
plt.legend(loc='lower right')
fig.set_size_inches(10, 5)
plt.savefig('.\\img\\missing.svg', format="svg")
```
