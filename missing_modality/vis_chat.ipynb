```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 13})
fig = plt.figure()

####### Missing Modality CHAT ########
x = np.array(["10%", "20%", "30%", "40%", "50%"])


y_nch = np.array([67.9, 64.8, 62.4, 59.9, 58.1])
e_nch = np.array([6.9, 4.8, 3.1, 2.4, 1.8])/2
plt.errorbar(x, y_nch, e_nch, label='CNN', marker = '', alpha=0.75)

y_nch = np.array([54.5, 53.7, 52.9, 52.2, 51.5])
e_nch = np.array([7.7, 5.9, 4.3, 3.3, 2.2])/2
plt.errorbar(x, y_nch, e_nch, label='CNN+LSTM', marker = '', alpha=0.75)

y_nch = np.array([82.4, 79.4, 76.3, 73.2, 69.8])
e_nch = np.array([3.2, 3.0, 2.6, 2.1, 2.0])/2
plt.errorbar(x, y_nch, e_nch, label='Transformer', marker = '', alpha=0.75)

y_nch = np.array([90.3, 88.2, 85.9, 82.8, 79.6])
e_nch = np.array([1.4, 1.4, 1.9, 2.0, 1.8])/2
plt.errorbar(x, y_nch, e_nch, label='Ours', marker = '', alpha=0.75)


plt.ylim([50,100])
plt.xlabel('Missingness Ratio')
plt.ylabel('AUROC')
plt.legend(loc='lower left')
fig.set_size_inches(6, 4)
plt.savefig('.\\img\\missing_chat.svg', format="svg")
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 13})
fig = plt.figure()

####### Noisy Modality CHAT ########
x = np.array(['10', "20", "30", "40", "50"])[::-1]

y_nch = np.array([64.7, 69.2, 70.7, 70.9, 70.9])[::-1]
e_nch = np.array([4.8, 6.5, 8.5, 9.0, 9.1])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='CNN', marker = '', alpha=0.75)

y_nch = np.array([51.9, 54.5, 60.8, 56.5, 56.3])[::-1]
e_nch = np.array([3.0, 8.4, 5.4, 9.6, 10.1])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='CNN+LSTM', marker = '', alpha=0.75)


y_nch = np.array([79.8, 80.5, 81.3, 83.6, 85.6])[::-1]
e_nch = np.array([3.2, 3.8, 3.7, 3.4, 3.0])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='Transformer', marker = '', alpha=0.75)


y_nch = np.array([88.1, 89.3, 91.6, 92.1, 92.1])[::-1]
e_nch = np.array([1.5, 0.9, 0.7, 1.2, 1.3])[::-1]/2
plt.errorbar(x, y_nch, e_nch, label='Ours', marker = '', alpha=0.75)


plt.ylim([50,100])
plt.xlabel('Signal to Noise Ratio (dB)')
plt.ylabel('AUROC')
plt.legend(loc='lower left')
fig.set_size_inches(6, 4)
plt.savefig('.\\img\\noisy_chat.svg', format="svg")
```
