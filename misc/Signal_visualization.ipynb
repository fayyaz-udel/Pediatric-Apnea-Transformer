```python
import mne
import matplotlib.pyplot as plt
from mne import make_fixed_length_events
import sleep_study as ss
import scipy
```

```python
def identity(df):
    return df

ss.__init__()
study = ss.data.study_list[110]
raw = ss.data.load_study(study, identity, verbose=True)
```

```python
sfreq = raw.info['sfreq']
tmax = 60.0 - 1. / sfreq

fixed_events = make_fixed_length_events(raw, id=0, duration=60.0, overlap=0.)
epochs = mne.Epochs(raw, fixed_events, event_id=[0], tmin=0, tmax=tmax, baseline=None, preload=True, proj=False,verbose=None)
```

```python
epochs.load_data()
data = epochs.get_data()
```
