```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()

x = np.array([5, 10, 20 ,30, 40, 50, '\u221e'])
y_chat = np.array([53.32,57.75,70.80,75.43,83.14,88.74,89.63])
yerr_chat = np.array([0.342,2.014,1.558,1.276,1.055,0.271,0.590 ])
plt.errorbar(x, y_chat, yerr=yerr_chat, label='CHAT')

y_nch = np.array([57.92,63.23,73.14,82.75,87.91,88.69,88.73])
yerr_nch = np.array([1.008,1.512,1.330,0.677,0.476,0.474,0.498 ])
plt.errorbar(x, y_nch, yerr=yerr_nch, label='NCH')

plt.ylim([50,100])
plt.xlabel('Signal to noise ratio')
plt.ylabel('AUROC')
plt.legend(loc='lower right')
plt.savefig('auroc.png', dpi=600)
```
